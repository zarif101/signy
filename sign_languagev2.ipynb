```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import cv2
import plaidml.keras
plaidml.keras.install_backend()
```

```python
train = pd.read_csv('sign_mnist_train.csv')
test = pd.read_csv('sign_mnist_test.csv')
```

```python
train.head()
```

```python
train_vals = train.iloc[:,1:].values
test_vals = test.iloc[:,1:].values
```

```python
y_train = train['label']
y_test = test['label']
```

```python
X_train = []
X_test = []
```

```python
for each in train_vals:
    new = np.reshape(each,(28,28))
    X_train.append(new)
    print('train added')
for each in test_vals:
    new = np.reshape(each,(28,28))
    X_test.append(new)
    print('test added')
```

```python
print(len(X_train))
print(len(X_test))
```

```python
X_train = np.array(X_train)
X_test = np.array(X_test)
```

```python
yes = X_train[0]/255
plt.imshow(yes)
```

```python
X_train_use = []
X_test_use = []
```

```python
for img in X_train:
    img = np.float32(img)
    ret,thresh = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
    X_train_use.append(thresh)
for img in X_test:
    img = np.float32(img)
    ret,thresh = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
    X_test_use.append(thresh)
```

```python
plt.imshow(X_train_use[2])
```

```python
X_train_use = np.array(X_train_use)
X_test_use = np.array(X_test_use)
```

```python
X_train_use = np.reshape(X_train_use,(27455,28,28,1))
```

```python
X_test_use = np.reshape(X_test_use,(7172,28,28,1))
```

```python
X_train_use[0].shape
```

```python
from keras.utils import to_categorical
```

```python
y_train_cat = to_categorical(y_train)
y_test_cat = to_categorical(y_test)
```

```python
#model time
```

```python
from keras.models import Sequential
from keras.layers import Conv2D,MaxPool2D,Dense,Flatten,Dropout
```

```python
model = Sequential()

model.add(Conv2D(filters=32,kernel_size=(3,3),input_shape=(28,28,1),activation='relu'))
model.add(Conv2D(filters=32,kernel_size=(3,3),activation='relu'))
model.add(MaxPool2D(pool_size=(2,2)))
model.add(Conv2D(filters=64,kernel_size=(3,3),activation='relu'))
model.add(Conv2D(filters=64,kernel_size=(3,3),activation='relu'))
model.add(MaxPool2D(pool_size=(2,2)))

model.add(Flatten())

model.add(Dense(units=128,activation='relu'))
model.add(Dense(units=64,activation='relu'))
model.add(Dropout(rate=0.4))
model.add(Dense(units=25,activation='softmax'))

model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
model.fit(X_train_use,y_train_cat,epochs=6)
```

```python
model.evaluate(X_test_use,y_test_cat)
```

```python
model.save('sign_language_thresh.h5')
```

```python
preds = model.predict_classes(X_test_use)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test.values,preds))
```
